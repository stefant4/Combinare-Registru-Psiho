```python
import pandas as pd

# Calea către fișierele Excel
base_path = r'C:\Users\stefa\Downloads\proiect Psihologie\partea 2\\'  
p21_file = f'{base_path}p21.xlsx'
p22_file = f'{base_path}p22.xlsx'
p23_file = f'{base_path}p23.xlsx'

# Citirea fișierelor Excel
p21 = pd.read_excel(p21_file)
p22 = pd.read_excel(p22_file)
p23 = pd.read_excel(p23_file)

# Verificarea coloanelor pentru a face merge pe toate coloanele identice
common_columns = p21.columns.intersection(p22.columns).intersection(p23.columns).tolist()

# Outer merge între p21 și p22 pe toate coloanele comune
merged_21_22 = pd.merge(p21, p22, on=common_columns, how='outer')

# Outer merge între rezultatul anterior și p23 pe toate coloanele comune
final_merged = pd.merge(merged_21_22, p23, on=common_columns, how='outer')

# Salvează rezultatul într-un nou fișier Excel
output_file = f'{base_path}merged_output.xlsx'
final_merged.to_excel(output_file, index=False)

print(f"Outer merge completat. Rezultatul este salvat în '{output_file}'.")
```
